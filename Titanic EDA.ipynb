```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
titanic=pd.read_csv('titanic.csv')
titanic.head()
```

```python
def age_group(age):
    if age <2:
        return 'Infant'
    elif age<18:
        return "Child"
    elif age<60:
        return 'Adult'
    else :
        return 'Senior'
    #added age_group column in dataset
titanic['age_group']=titanic['Age'].map(age_group)
titanic['Age'].map(age_group)
titanic['Age'].map(lambda x: age_group(x))
AgeSurvived=titanic.pivot_table(index='age_group', values='Survived', aggfunc='sum')
print(AgeSurvived)
plt.title('Survival by Age Group',fontsize=16,color='black')
ax=plt.bar(AgeSurvived.index,AgeSurvived['Survived'])
```

```python
titanic.groupby(["Pclass"])["Survived"].value_counts(normalize=True).unstack().plot(kind="bar", stacked=True, )
plt.xlabel("Types of passenger classes")
plt.ylabel("% of survivals")
plt.title("Survival of passengers based on the Pclass")
```

```python
titanic.groupby(["Sex"])["Survived"].value_counts(normalize=True).unstack().plot(kind="bar", stacked=True, )
plt.xlabel("Types of passenger classes")
plt.ylabel("% of survivals")
plt.title("Survival of passengers based on the Pclass")
```

```python
titanic.groupby(["Sex", "Pclass"])["Survived"].value_counts(normalize=True).unstack().plot(kind="bar", stacked=True, )
plt.xlabel("Types of passenger classes")
plt.ylabel("% of survivals")
plt.title("Survival of passengers based on the Pclass")
```

```python
titanic.groupby(["age_group",'Pclass','Sex'])["Survived"].value_counts(normalize=True).unstack().plot(kind="bar", stacked=True, )
plt.xlabel("Types of passenger classes")
plt.ylabel("% of survivals")
plt.title("Survival of passengers based on the Pclass")
```
